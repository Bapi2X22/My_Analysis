```python
import ROOT
%jsroot on

# Create histograms
h_proc = ROOT.TH1F("h_proc", "Lead pT; pT [GeV]; Events", 100, 0, 500)
h_raw  = ROOT.TH1F("h_raw",  "Lead pT; pT [GeV]; Events", 100, 0, 500)

# Fill histograms (assuming proc_lead_pt and raw_lead_pt[0] are iterable, e.g. numpy arrays or lists)
for val in proc_lead_pt:
    h_proc.Fill(val)

data = np.asarray(raw_lead_pt)
data = data[~np.isnan(data)]  # Remove NaN values

for val in data:
    h_raw.Fill(val)

# Disable stats box
h_proc.SetStats(0)
h_raw.SetStats(0)

# Calculate stats
proc_entries = h_proc.GetEntries()
proc_mean    = h_proc.GetMean()
proc_std     = h_proc.GetRMS()

raw_entries  = h_raw.GetEntries()
raw_mean     = h_raw.GetMean()
raw_std      = h_raw.GetRMS()

# Legend with smaller font
legend = ROOT.TLegend(0.55, 0.65, 0.88, 0.85)  # (x1, y1, x2, y2)
legend.SetTextSize(0.025)  # smaller text
legend.SetBorderSize(0)
legend.SetFillStyle(0)

# Add entries with formatted stats
legend.AddEntry(h_proc,
                f"Processed pT: N={int(proc_entries)}, ={proc_mean:.1f}, ={proc_std:.1f}",
                "l")
legend.AddEntry(h_raw,
                f"Raw pT: N={int(raw_entries)}, ={raw_mean:.1f}, ={raw_std:.1f}",
                "l")

# Draw again
c = ROOT.TCanvas("c", "c", 800, 600)
h_proc.SetLineColor(ROOT.kRed)
h_raw.SetLineColor(ROOT.kBlue)

# Adjust y-axis
ymax = max(h_proc.GetMaximum(), h_raw.GetMaximum())
h_proc.SetMaximum(ymax * 1.5)
h_proc.SetMinimum(0)

h_proc.Draw("HIST")
h_raw.Draw("HIST SAME")

legend.Draw()
# c.SetLogy()
c.Draw()
c.SaveAs("lead_pt_comparison.png")


```
